### 1. Carregar a base TITANIC do seaborn

```python
from seaborn import load_dataset

# carregar base de dados
df = load_dataset('titanic')
df
```

### 2. Observar a relação dos atributos "pclass" com "class"

```python
df[["pclass", "class"]]
```

```python
df = df.drop(columns=["class"])
df
```

### 3. Observar a quantidade de NaN existente no atributo "deck"

```python
df["deck"].isna()
```

```python
df = df.drop(columns=["deck", "alive", "adult_male", "sex", "embarked"])
df
```

### 4. Remover as linhas que tenham valores NaN

```python
df = df.dropna()
df
```

```python
df.info()
```

### 5. Transformar todos os atributos categóricos usando One-Hot Encoder

```python
import pandas as pd
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder

column_transformer = make_column_transformer((OneHotEncoder(), ['who', 'embark_town', 'alone']), remainder='passthrough')

transformed_data = column_transformer.fit_transform(df)
columns_names = column_transformer.get_feature_names_out()
columns_names
```

```python
df_final = pd.DataFrame(data=transformed_data, columns=columns_names)
df_final.columns = ['is_child', 'is_man', 'is_woman', 'Cherboug_as_embarked_town', 'Queenstown_as_embarked_town', 'Southampton_as_embarked_town', 'not_alone', 'alone', 'survived', 'pclass', 'age', 'sibsp', 'parch', 'fare']
df_final = df_final.drop(['not_alone'], axis=1)
df_final
```

### 6. Coluna target é a "survived"

```python
y = df_final['survived'] # dados que queremos prever
X = df_final.drop(labels=['survived'], axis=1) # dados que usaremos para prever
```

```python
X
```

### 7. Separar os dados de treinamento e testes

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=None, stratify=y)
```

```python
X_train
```

```python
X_test
```

### 8. Treinar 2 árvores de decisão com características diferentes

```python
from sklearn.tree import DecisionTreeClassifier

tree_1 = DecisionTreeClassifier(criterion='gini', max_depth=4)
tree_2 = DecisionTreeClassifier(criterion='entropy', max_depth=2)
```

```python
tree_1.fit(X_train, y_train)
```

```python
tree_2.fit(X_train, y_train)
```

### 9. Treinar 2 kNN com características diferentes

```python
from sklearn.neighbors import KNeighborsClassifier

knn_1 = KNeighborsClassifier(n_neighbors=5, metric='euclidean', algorithm='brute')
knn_2 = KNeighborsClassifier(n_neighbors=7, metric='minkowski', algorithm='kd_tree')
```

```python
knn_1.fit(X_train, y_train)
```

```python
knn_2.fit(X_train, y_train)
```

```python
from sklearn import metrics

tree_result_1 = tree_1.predict(X_test)
tree_result_2 = tree_2.predict(X_test)
knn_result_1 = knn_1.predict(X_test)
knn_result_2 = knn_2.predict(X_test)

acc1 = metrics.accuracy_score(tree_result_1, y_test)
show1 = round(acc1 * 100)
print("DecisionTreeClassifier(criterion='gini', max_depth=4) -> {}%".format(show1))

acc2 = metrics.accuracy_score(tree_result_2, y_test)
show2 = round(acc2 * 100)
print("DecisionTreeClassifier(criterion='entropy', max_depth=2) -> {}%".format(show2))

acc3 = metrics.accuracy_score(knn_result_1, y_test)
show3 = round(acc3 * 100)
print("KNeighborsClassifier(algorithm='brute', metric='euclidean', n_neighbors=5) -> {}%".format(show3))

acc4 = metrics.accuracy_score(knn_result_1, y_test)
show4 = round(acc4 * 100)
print("KNeighborsClassifier(algorithm='kd_tree', metric='minkowski', n_neighbors=7) -> {}%".format(show4))
```
